```python
import pandas as pd
import numpy as np
import os
import glob
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Pfad zur CSV-Datei
file_path = '../participant_package/reefer_release.csv'

# Lade die Daten in einen DataFrame
df_reefer = pd.read_csv(file_path, sep=';', decimal=',')

# Zeige die ersten paar Zeilen an
df_reefer.head(50)
```

```python
# 1. Leerzeichen am Anfang/Ende entfernen und Leerzeichen durch Unterstriche ersetzen
df_reefer.columns = df_reefer.columns.str.strip().str.replace(' ', '_')

# 2. Alles kleinschreiben (optional, aber extrem hilfreich für schnelles Tippen)
df_reefer.columns = df_reefer.columns.str.lower()

# Kurzer Check der neuen Namen
print("Bereinigte Spaltennamen in df_reefer:")
print(df_reefer.columns.tolist())
```

```python
# 1. Grundstruktur und Datentypen
print("--- Info ---")
print(df_reefer.info())
```

```python
# 2. Wie viele Missings pro Spalte?
print("\n--- Fehlende Werte ---")
missings = df_reefer.isna().sum()
print(missings[missings > 0]) # Zeigt nur Spalten mit Fehlern
```

```python
#Container Size missings lösen


df_reefer['ContainerSize'] = df_reefer.groupby('container_visit_uuid')['containersize'].transform(lambda x: x.ffill().bfill())

# Kurzer Check: Wie viele Missings sind jetzt noch übrig?
remaining_missings = df_reefer['ContainerSize'].isna().sum()
print(f"Missings in ContainerSize nach der Korrektur: {remaining_missings}")

# Falls noch Missings übrig sind, bedeutet das, dass für diese UIDs 
# in KEINER Zeile eine Größe hinterlegt war.
```

```python
# 1. Zeilen löschen, bei denen 'containersize' fehlt
df_reefer = df_reefer.dropna(subset=['containersize'])

# 2. Missings in 'stack_tier' durch den Wert 1 ersetzen
df_reefer['stack_tier'] = df_reefer['stack_tier'].fillna(1)

# Kurze Erfolgskontrolle
print(f"✅ Zeilen mit fehlender Größe gelöscht. Neue Zeilenanzahl: {len(df_reefer)}")
print(f"✅ Verbleibende Missings in 'stack_tier': {df_reefer['stack_tier'].isna().sum()}")

# Optional: Prüfen, ob stack_tier jetzt den richtigen Datentyp hat (z.B. Integer)
# df_reefer['stack_tier'] = df_reefer['stack_tier'].astype(int)
```

```python
# Berechnung der Missings
missing_count = df_reefer.isna().sum()
missing_percentage = (missing_count / len(df_reefer)) * 100

# Zusammenführen in eine schöne Tabelle
missing_report = pd.DataFrame({
    'Anzahl Missings': missing_count,
    'Anteil (%)': missing_percentage.round(2)
})

# Nur Spalten anzeigen, die tatsächlich Missings haben, sortiert nach Wichtigkeit
missing_report = missing_report[missing_report['Anzahl Missings'] > 0].sort_values(by='Anteil (%)', ascending=False)

print(f"Gesamtanzahl Zeilen im Datensatz: {len(df_reefer)}")
print("-" * 40)
if missing_report.empty:
    print("✅ Wahnsinn! Der Datensatz ist jetzt komplett sauber (keine Missings).")
else:
    print(missing_report)
```

```python
import openmeteo_requests

import pandas as pd
import requests_cache
from retry_requests import retry

# Setup the Open-Meteo API client with cache and retry on error
cache_session = requests_cache.CachedSession('.cache', expire_after = 3600)
retry_session = retry(cache_session, retries = 5, backoff_factor = 0.2)
openmeteo = openmeteo_requests.Client(session = retry_session)

# Make sure all required weather variables are listed here
# The order of variables in hourly or daily is important to assign them correctly below
url = "https://historical-forecast-api.open-meteo.com/v1/forecast"
params = {
	"latitude": 53.5507,
	"longitude": 9.993,
	"start_date": "2025-01-01",
	"end_date": "2026-03-31",
	"hourly": ["temperature_2m", "precipitation", "wind_speed_10m", "wind_direction_10m", "cloud_cover"],
	"timezone": "Europe/Berlin",
}
responses = openmeteo.weather_api(url, params = params)

# Process first location. Add a for-loop for multiple locations or weather models
response = responses[0]
print(f"Coordinates: {response.Latitude()}°N {response.Longitude()}°E")
print(f"Elevation: {response.Elevation()} m asl")
print(f"Timezone: {response.Timezone()}{response.TimezoneAbbreviation()}")
print(f"Timezone difference to GMT+0: {response.UtcOffsetSeconds()}s")

# Process hourly data. The order of variables needs to be the same as requested.
hourly = response.Hourly()
hourly_temperature_2m = hourly.Variables(0).ValuesAsNumpy()
hourly_precipitation = hourly.Variables(1).ValuesAsNumpy()
hourly_wind_speed_10m = hourly.Variables(2).ValuesAsNumpy()
hourly_wind_direction_10m = hourly.Variables(3).ValuesAsNumpy()
hourly_cloud_cover = hourly.Variables(4).ValuesAsNumpy()

hourly_data = {"date": pd.date_range(
	start = pd.to_datetime(hourly.Time() + response.UtcOffsetSeconds(), unit = "s", utc = True),
	end =  pd.to_datetime(hourly.TimeEnd() + response.UtcOffsetSeconds(), unit = "s", utc = True),
	freq = pd.Timedelta(seconds = hourly.Interval()),
	inclusive = "left"
)}

hourly_data["temperature_2m"] = hourly_temperature_2m
hourly_data["precipitation"] = hourly_precipitation
hourly_data["wind_speed_10m"] = hourly_wind_speed_10m
hourly_data["wind_direction_10m"] = hourly_wind_direction_10m
hourly_data["cloud_cover"] = hourly_cloud_cover

hourly_dataframe = pd.DataFrame(data = hourly_data)
hourly_dataframe
```

```python
# 1. Datetime-Format erzwingen UND Zeitzonen-Info entfernen (.dt.tz_localize(None))
df_reefer['eventtime'] = pd.to_datetime(df_reefer['eventtime']).dt.tz_localize(None)
hourly_dataframe['date'] = pd.to_datetime(hourly_dataframe['date']).dt.tz_localize(None)

# 2. Der Merge (jetzt ohne Fehler)
df_reefer_combined = pd.merge(
    df_reefer, 
    hourly_dataframe, 
    left_on='eventtime', 
    right_on='date', 
    how='left'
)

# 3. Aufräumen
df_reefer_combined = df_reefer_combined.drop(columns=['date'], errors='ignore')

print(f"✅ Erfolg! Datensatz kombiniert. Spalten: {df_reefer_combined.columns.tolist()}")
```

```python
# 1. Sicherstellen, dass beide Zeitspalten im Datetime-Format vorliegen
df_reefer['eventtime'] = pd.to_datetime(df_reefer['eventtime'])
hourly_dataframe['date'] = pd.to_datetime(hourly_dataframe['date'])

# 2. Die Zusammenführung (Merge)
# Wir nutzen left_on und right_on, da die Spalten unterschiedlich heißen.
df_reefer_combined = pd.merge(
    df_reefer, 
    hourly_dataframe, 
    left_on='eventtime', 
    right_on='date', 
    how='left'
)

# 3. Die redundante Spalte 'date' löschen (da wir eventtime behalten)
df_reefer_combined = df_reefer_combined.drop(columns=['date'], errors='ignore')

# 4. Check des Ergebnisses
print(f"Neuer Datensatz erstellt mit {df_reefer_combined.shape[1]} Spalten.")
df_reefer_combined
```
