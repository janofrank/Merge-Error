```python
import pandas as pd
import numpy as np
import os
import glob
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
############## DATEI LADEN ##############

# Pfad zur CSV-Datei
file_path = '../participant_package/reefer_release.csv'

# Lade die Daten in einen DataFrame
df_reefer = pd.read_csv(file_path, sep=';', decimal=',')

# Zeige die ersten paar Zeilen an
df_reefer.head(50)
```

```python
############## Spaltennamen von Leerzeichen befreien ##############

# 1. Leerzeichen am Anfang/Ende entfernen und Leerzeichen durch Unterstriche ersetzen
df_reefer.columns = df_reefer.columns.str.strip().str.replace(' ', '_')

# 2. Alles kleinschreiben (optional, aber extrem hilfreich für schnelles Tippen)
df_reefer.columns = df_reefer.columns.str.lower()

# Kurzer Check der neuen Namen
print("Bereinigte Spaltennamen in df_reefer:")
print(df_reefer.columns.tolist())
```

```python
# 1. Grundstruktur und Datentypen
print("--- Info ---")
print(df_reefer.info())
```

```python
# 2. Wie viele Missings pro Spalte?
print("\n--- Fehlende Werte ---")
missings = df_reefer.isna().sum()
print(missings[missings > 0]) # Zeigt nur Spalten mit Fehlern
```

```python
############## Zeilen löschen, bei denen 'containersize' fehlt ##############


df_reefer = df_reefer.dropna(subset=['containersize'])

# 2. Missings in 'stack_tier' durch den Wert 1 ersetzen
df_reefer['stack_tier'] = df_reefer['stack_tier'].fillna(1)

# Kurze Erfolgskontrolle
print(f"✅ Zeilen mit fehlender Größe gelöscht. Neue Zeilenanzahl: {len(df_reefer)}")
print(f"✅ Verbleibende Missings in 'stack_tier': {df_reefer['stack_tier'].isna().sum()}")

# Optional: Prüfen, ob stack_tier jetzt den richtigen Datentyp hat (z.B. Integer)
# df_reefer['stack_tier'] = df_reefer['stack_tier'].astype(int)
```

```python
# Sicherstellen, dass eventtime als Datetime vorliegt
df_reefer['eventtime'] = pd.to_datetime(df_reefer['eventtime'])

# Auf volle Stunde runden, damit wirklich stündlich gruppiert wird
df_reefer['hour_timestamp'] = df_reefer['eventtime'].dt.floor('h')

# Datum für die Tagesbetrachtung ableiten
df_reefer['date'] = df_reefer['hour_timestamp'].dt.date

# 1) Über alle Container hinweg den Gesamtverbrauch pro Stunde berechnen
hourly_energy = (
    df_reefer
    .groupby(['date', 'hour_timestamp'], as_index=False)['ttlenergyconshour']
    .sum()
    .rename(columns={'ttlenergyconshour': 'total_energy_hour'})
)

# 2) Pro Tag den höchsten stündlichen Gesamtverbrauch bestimmen
hourly_energy['daily_max_energy'] = (
    hourly_energy
    .groupby('date')['total_energy_hour']
    .transform('max')
)

# 3) Neue Kennzeichnung:
#    1 = diese Stunde ist die Peak-Stunde des Tages
#    0 = diese Stunde ist nicht die Peak-Stunde des Tages
hourly_energy['is_energy_peak_hour'] = (
    hourly_energy['total_energy_hour'] == hourly_energy['daily_max_energy']
).astype(int)

# 4) Die Peak-Kennzeichnung zurück an den ursprünglichen DataFrame anhängen
df_reefer = df_reefer.merge(
    hourly_energy[['hour_timestamp', 'is_energy_peak_hour']],
    on='hour_timestamp',
    how='left'
)

# 5) Hilfsspalten wieder entfernen, falls du sie nicht behalten möchtest
df_reefer.drop(columns=['hour_timestamp', 'date'], inplace=True)

# Kurze Kontrolle
print(df_reefer[['eventtime', 'ttlenergyconshour', 'is_energy_peak_hour']].head())
df_reefer.head()
```

```python
import openmeteo_requests

import pandas as pd
import requests_cache
from retry_requests import retry

# Setup the Open-Meteo API client with cache and retry on error
cache_session = requests_cache.CachedSession('.cache', expire_after = 3600)
retry_session = retry(cache_session, retries = 5, backoff_factor = 0.2)
openmeteo = openmeteo_requests.Client(session = retry_session)

# Make sure all required weather variables are listed here
# The order of variables in hourly or daily is important to assign them correctly below
url = "https://historical-forecast-api.open-meteo.com/v1/forecast"
params = {
	"latitude": 53.5507,
	"longitude": 9.993,
	"start_date": "2025-01-01",
	"end_date": "2026-03-31",
	"hourly": ["temperature_2m", "precipitation", "wind_speed_10m", "wind_direction_10m", "cloud_cover"],
	"timezone": "Europe/Berlin",
}
responses = openmeteo.weather_api(url, params = params)

# Process first location. Add a for-loop for multiple locations or weather models
response = responses[0]
print(f"Coordinates: {response.Latitude()}°N {response.Longitude()}°E")
print(f"Elevation: {response.Elevation()} m asl")
print(f"Timezone: {response.Timezone()}{response.TimezoneAbbreviation()}")
print(f"Timezone difference to GMT+0: {response.UtcOffsetSeconds()}s")

# Process hourly data. The order of variables needs to be the same as requested.
hourly = response.Hourly()
hourly_temperature_2m = hourly.Variables(0).ValuesAsNumpy()
hourly_precipitation = hourly.Variables(1).ValuesAsNumpy()
hourly_wind_speed_10m = hourly.Variables(2).ValuesAsNumpy()
hourly_wind_direction_10m = hourly.Variables(3).ValuesAsNumpy()
hourly_cloud_cover = hourly.Variables(4).ValuesAsNumpy()

hourly_data = {"date": pd.date_range(
	start = pd.to_datetime(hourly.Time() + response.UtcOffsetSeconds(), unit = "s", utc = True),
	end =  pd.to_datetime(hourly.TimeEnd() + response.UtcOffsetSeconds(), unit = "s", utc = True),
	freq = pd.Timedelta(seconds = hourly.Interval()),
	inclusive = "left"
)}

hourly_data["temperature_2m"] = hourly_temperature_2m
hourly_data["precipitation"] = hourly_precipitation
hourly_data["wind_speed_10m"] = hourly_wind_speed_10m
hourly_data["wind_direction_10m"] = hourly_wind_direction_10m
hourly_data["cloud_cover"] = hourly_cloud_cover

hourly_dataframe = pd.DataFrame(data = hourly_data)
hourly_dataframe
```

```python
import xgboost as xgb
import optuna
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_absolute_error
import warnings

# Warnungen unterdrücken für sauberere Konsolenausgabe in der Zelle
warnings.filterwarnings('ignore')

def compute_pinball_loss(y_true, y_pred, alpha=0.9):
    """Berechnet den Pinball Loss (Quantile Loss)."""
    residuals = y_true - y_pred
    return np.mean(np.maximum(alpha * residuals, (alpha - 1) * residuals))

# ==============================================================================
# 1. Datenvorbereitung & Aggregation
# ==============================================================================
print("1. Aggregiere bereits geladene 'df_reefer_combined' Container-Daten...")

# Stelle sicher, dass eventtime datetime ist und entferne eventuelle Zeitzonen (tz-naive),
# um Union-Errors ("Cannot compare tz-naive and tz-aware timestamps") zu vermeiden
df_reefer_combined['eventtime'] = pd.to_datetime(df_reefer_combined['eventtime']).dt.tz_localize(None)

# Finde Wetter- und Peak-Features im vorhandenen Dataframe
potential_weather_cols = ['temperature_2m', 'precipitation', 'wind_speed_10m', 
                          'wind_direction_10m', 'cloud_cover', 'is_energy_peak_hour']
weather_cols = [col for col in potential_weather_cols if col in df_reefer_combined.columns]

# Aggregation: Gesamtstromverbrauch summieren, anliegendes Wetter beibehalten (max/first)
agg_funcs = {'avpowercons': 'sum'}
for col in weather_cols:
    agg_funcs[col] = 'max' 

df_agg = df_reefer_combined.groupby('eventtime').agg(agg_funcs).reset_index()

# Sortieren und sicherstellen, dass wir eine lückenfreie stündliche Zeitreihe haben
df_agg = df_agg.sort_values('eventtime').set_index('eventtime')
df_agg = df_agg.resample('1h').asfreq()

# Gesamtlast in kW ableiten und alte Spalte entfernen
df_agg['target_load_kw'] = df_agg['avpowercons'] / 1000.0
df_agg.drop(columns=['avpowercons'], inplace=True)

# Lücken im Wetter forward-fillen (Target darf nicht manipuliert werden)
df_agg[weather_cols] = df_agg[weather_cols].ffill()

# ==============================================================================
# 2. Feature Engineering (Strikte .shift(24) Logik)
# ==============================================================================
print("2. Führe Feature Engineering durch (Shift-Logik)...")
df_features = df_agg[['target_load_kw']].copy()

if 'is_energy_peak_hour' in df_agg.columns:
    df_features['is_energy_peak_hour'] = df_agg['is_energy_peak_hour']
    
# a) Zeitbasierte Features für den Ziel-Zeitstempel T
df_features['hour'] = df_features.index.hour
df_features['dayofweek'] = df_features.index.dayofweek
df_features['month'] = df_features.index.month
df_features['is_weekend'] = df_features.index.dayofweek.isin([5, 6]).astype(int)

# b) Historische Lastwerte um exakt 24, 48 und 168 Stunden shiften
df_features['lag_24'] = df_agg['target_load_kw'].shift(24)
df_features['lag_48'] = df_agg['target_load_kw'].shift(48)
df_features['lag_168'] = df_agg['target_load_kw'].shift(168)

# c) Wetterdaten um 24 Stunden shiften!
# Grund: Das exakte 24h-Wetter-Forecast für T steht in Zeile T-24
for weather_feature in [col for col in weather_cols if col != 'is_energy_peak_hour']:
    df_features[f"{weather_feature}_forecast"] = df_agg[weather_feature].shift(24)
    
# d) Rolling-Window Features basierend auf dem ersten verfügbaren Lag (lag_24)
df_features['rolling_mean_24_on_lag24'] = df_features['lag_24'].rolling(window=24).mean()
df_features['rolling_std_24_on_lag24'] = df_features['lag_24'].rolling(window=24).std()

# Features auswählen: Die echte Last ("target_load_kw") und "is_energy_peak_hour" sind keine Eingabefeatures
exclude_from_training = ['target_load_kw', 'is_energy_peak_hour']
features_to_use = [col for col in df_features.columns if col not in exclude_from_training]

# Droppe Zeilen mit initialen NaNs (entstanden durch Shifts/Rollings)
df_final = df_features.dropna(subset=features_to_use + ['target_load_kw']).copy()

# ==============================================================================
# 3. Validierungs-Split definieren
# ==============================================================================
print("3. Erstelle Validierungssplit (Testset: Jan 2026, Modell trainiert auf allem davor)...")
train_mask = df_final.index < '2026-01-01'
test_mask = (df_final.index >= '2026-01-01') & (df_final.index < '2026-02-01')

df_train = df_final[train_mask]
df_test = df_final[test_mask]

X_train, y_train = df_train[features_to_use], df_train['target_load_kw']
X_test, y_test = df_test[features_to_use], df_test['target_load_kw']

# ==============================================================================
# 4. Optuna Hyperparameter-Tuning
# ==============================================================================
print("4. Starte Optuna Hyperparameter-Tuning (TimeSeriesSplit, nur auf Trainingsdaten)...")

def optuna_objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 1000),
        'max_depth': trial.suggest_int('max_depth', 1, 15),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.3, 1.0),
        'objective': 'reg:absoluteerror',
        'random_state': 42,
        'n_jobs': -1
    }
    
    tscv = TimeSeriesSplit(n_splits=3)
    cv_scores = []
    
    for train_idx, val_idx in tscv.split(X_train):
        X_tr, X_val = X_train.iloc[train_idx], X_train.iloc[val_idx]
        y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[val_idx]
        
        regr = xgb.XGBRegressor(**params)
        regr.fit(X_tr, y_tr, verbose=False)
        cv_scores.append(mean_absolute_error(y_val, regr.predict(X_val)))
        
    return np.mean(cv_scores)

# Reduzierte n_trials auf 15, um nicht zu lange im Notebook zu hängen (kann stark erhöht werden: zB 100)
study = optuna.create_study(direction='minimize')
study.optimize(optuna_objective, n_trials=100)
best_params = study.best_params
print(f"   -> Beste Parameter ermittelt: {best_params}")

# ==============================================================================
# 5. Finale Modelle Trainieren & Test-Set Evaluierung
# ==============================================================================
print("5. Trainiere finale Modelle und evaluiere auf dem Januar Testset...")

# Model 1: Point Forecast (optimiert auf MAE)
model1_params = best_params.copy()
model1_params['objective'] = 'reg:absoluteerror'
reg_m1 = xgb.XGBRegressor(**model1_params, random_state=42, n_jobs=-1)
reg_m1.fit(X_train, y_train)

# Model 2: 90% Quantile Forecast (optimiert auf Pinball Loss, P90)
model2_params = best_params.copy()
model2_params['objective'] = 'reg:quantileerror'
model2_params['quantile_alpha'] = 0.9
reg_m2 = xgb.XGBRegressor(**model2_params, random_state=42, n_jobs=-1)
reg_m2.fit(X_train, y_train)

# Test-Set Inferenz
preds_m1_test = reg_m1.predict(X_test)
preds_m2_test = reg_m2.predict(X_test)

# Post-Processing: 
# 1. Negative Werte clippen
# 2. Regel anwenden: P90 Quantil muss größer/gleich als Punktvorhersage sein
preds_m1_test = np.maximum(0, preds_m1_test)
preds_m2_test = np.maximum(0, preds_m2_test)
preds_m2_test = np.maximum(preds_m2_test, preds_m1_test)

# Metriken Berechnen
mae_all = mean_absolute_error(y_test, preds_m1_test)
pb_90 = compute_pinball_loss(y_test, preds_m2_test, alpha=0.9)

if 'is_energy_peak_hour' in df_test.columns:
    peak_mask = df_test['is_energy_peak_hour'] == 1
else:
    # Falls das Feature fehlt: Wir generieren einen dynamischen 10% Peak Fallback
    peak_mask = y_test >= y_test.quantile(0.9)
    
mae_peak = mean_absolute_error(y_test[peak_mask], preds_m1_test[peak_mask]) if peak_mask.sum() > 0 else 0
final_score = 0.5 * mae_all + 0.3 * mae_peak + 0.2 * pb_90

print()
print("========= ERGEBNISSE JANUAR 2026 =========")
print(f" MAE All:        {mae_all:.4f}")
print(f" MAE Peak:       {mae_peak:.4f}")
print(f" Pinball (P90):  {pb_90:.4f}")
print(f" => FINAL SCORE: {final_score:.4f}")
print("==========================================")

# ==============================================================================
# 6. Target-Timestamps iterieren und Vorhersage in CSV speichern
# ==============================================================================
print("6. Generiere Vorhersagen für die finalen Target-Zeitstempel...")
target_timestamps_path = "target_timestamps.csv"

# Target CSV in Index laden und Zeitzone entfernen für den nahtlosen Vergleich
df_targets = pd.read_csv(target_timestamps_path)
df_targets['timestamp_utc'] = pd.to_datetime(df_targets['timestamp_utc']).dt.tz_localize(None)
targets = df_targets['timestamp_utc'].sort_values().drop_duplicates()

# Lückenfreie Erweiterung des Feature-Indexes, damit die Lag-Funktion durchläuft
all_times = df_agg.index.union(targets).sort_values().drop_duplicates()
df_ext = df_agg.reindex(all_times)

# Neues dynamisches Feature Engineering auf den erweiterten Zeitstempeln
df_features_ext = pd.DataFrame(index=df_ext.index)
df_features_ext['hour'] = df_features_ext.index.hour
df_features_ext['dayofweek'] = df_features_ext.index.dayofweek
df_features_ext['month'] = df_features_ext.index.month
df_features_ext['is_weekend'] = df_features_ext.index.dayofweek.isin([5, 6]).astype(int)

df_features_ext['lag_24'] = df_ext['target_load_kw'].shift(24)
df_features_ext['lag_48'] = df_ext['target_load_kw'].shift(48)
df_features_ext['lag_168'] = df_ext['target_load_kw'].shift(168)

for weather_feature in [col for col in weather_cols if col != 'is_energy_peak_hour']:
    df_features_ext[f"{weather_feature}_forecast"] = df_ext[weather_feature].shift(24)
    
df_features_ext['rolling_mean_24_on_lag24'] = df_features_ext['lag_24'].rolling(window=24).mean()
df_features_ext['rolling_std_24_on_lag24'] = df_features_ext['lag_24'].rolling(window=24).std()

# Filter auf die exakten Stunden die wir für die finale Vorhersage brauchen
X_infer = df_features_ext.loc[targets, features_to_use].copy()

# Vorhersagen
preds_sub_power = reg_m1.predict(X_infer)
preds_sub_p90 = reg_m2.predict(X_infer)

# Post-Processing
preds_sub_power = np.maximum(0, preds_sub_power)
preds_sub_p90 = np.maximum(0, preds_sub_p90)
preds_sub_p90 = np.maximum(preds_sub_p90, preds_sub_power)

df_submission = pd.DataFrame({
    'timestamp_utc': targets,
    'pred_power_kw': preds_sub_power,
    'pred_p90_kw': preds_sub_p90
})

# Programmatisch Duplikate abfangen: Nimmt automatisch den ersten Hit
df_submission = df_submission.groupby('timestamp_utc', as_index=False).first().reset_index(drop=True)

# Export
df_submission.to_csv('predictions.csv', index=False)
print("Erfolgreich abgeschlossen! Datei 'predictions.csv' wurde gespeichert.")
```
